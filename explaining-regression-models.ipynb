### Regression
**Regression** is a supervised machine learning process. It is similar to classification, but instead of predicting a label (target), it tries to predict a continuous (numeric) value.<br><br>
The fact is that **sklearn** is capable of applying many of the same **classification** models to **regression** problems. In effect, the API is the same and calls *.fit*, *.score*, and *.predict*.<br><br>
For the **regression**, we will use a Boston housing dataset.

### Reading the Boston Housing Dataset

```python
# loading boston dataset
b = load_boston()
```

### Preparing the dataset

```python
# creating X dataset
bos_X = pd.DataFrame(b.data, columns=b.feature_names)

# creating y dataset
bos_y = b.target

# train_test_split dataset
bos_X_train, bos_X_test, bos_y_train, bos_y_test = model_selection.train_test_split(
    bos_X,
    bos_y,
    test_size=0.3,
    random_state=42
)

bos_X
```

#### We will evaluate the results of a regression using XGBoost model as a base

```python
xgr = xgb.XGBRegressor(
    random_state=42, base_score=0.5
)

xgr.fit(bos_X_train, bos_y_train)
```

### Shapley
It is important to understand that the **Shapley** lib does not depend on the model. This lib also provides global insights into our model and helps explain individual predictions. Widely used in black box models.
Initially, we will see the prediction for index 5.

```python
sample_idx = 5

xgr.predict(bos_X.iloc[[sample_idx]])
```

To use the model, we need to create a *TreeExplainer* from our model and estimate the **SHAP** values for our samples. To use **SHAP** in Jupyter, you need to call the *initjs* function

```python
shap.initjs()

exp = shap.TreeExplainer(xgr)

vals = exp.shap_values(bos_X)
```

With the explainer and the **SHAP** values, we can create a force plot to explain the prediction. This chart tells you that the base forecast is 23, and that population status (LSAT) and property tax rate (TAX) push the price up, while the number of rooms pushes the price down

```python
shap.force_plot(
    exp.expected_value,
    vals[sample_idx],
    bos_X.iloc[[sample_idx]]
)
```

We can visualize the force plot for all samples, as well as get a general idea of ​​the behavior

```python
shap.force_plot(
    exp.expected_value, vals, bos_X
)

#plt.savefig("images/explaining_regression_models_force_plot_shap", dpi=700)
#plt.show()
```

From this sample strength chart, we see that the **LSTAT** attribute has a big impact. To visualize how **LSTAT** affects the result, we can create a dependency graph. The lib will automatically choose an attribute to color accordingly.<br><br>
From the dependency graph for **LSTAT**, we can notice that as **LSTAT** increases (percentage of lowest status population), the **SHAP** value decreases (forcing the target downward ). We can also see that, as the rate decreases, the **SHAP** value increases.

```python
fig, ax = plt.subplots(figsize=(6, 4))

shap.dependence_plot("LSTAT", vals, bos_X)

fig.savefig("images/explaining_regression_models_lstat_dependence_plot_shap", bbox_inches="tight")
```

Next, we will create another dependency graph, to explore the **DIS** (distance to employment centers) attribute. Apparently, this attribute has little effect

```python
fig, ax = plt.subplots(figsize=(6, 4))

shap.dependence_plot("DIS", vals, bos_X, interaction_index="RM")

fig.savefig("images/explaining_regression_models_dis_dependence_plot_shap", bbox_inches="tight")
```
