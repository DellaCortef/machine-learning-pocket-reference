### Reading the Titanic Dataset

```python
df = pd.read_csv("titanic_dataset.csv", index_col=0)
df.head(5)
```

### Deleting _Class_ feature at the end
We are deleting because is the same as _pclass_ (same result, same data)

```python
df = df.drop('Class', axis = 'columns')
df
```

### Converting DataFrame Column Names to Lowercase snakecase

```python
df.columns = (df.columns
                .str.replace('(?<=[a-z])(?=[A-Z])', '_', regex=True)
                .str.lower()
             )

df
```

### Preparing the dataset

```python
# dropping columns taht do not add value
df = df.drop(columns = ['name',
                        'name_wiki',
                        'wiki_id',
                        'hometown',
                        'destination',
                        'ticket',
                        'lifeboat',
                        'body',
                        'cabin',
                        'age'])

# using get_dummies function to convert object to int
df = pd.get_dummies(df)

# dropping redundant features
df = df.drop(columns = ['sex_male'])

#remove rows with any values that are not finite (NaN or infite)
df = df[np.isfinite(df).all(1)]

# first, we need to create a series of the target feature
y = df.survived

# then, we create a DataFrame with the attributes
X = df.drop(columns = ['survived'])

# using the scikit-learn to split 30% to test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
X.isnull().sum()
```

```python
dt = DecisionTreeClassifier(
    random_state=42, max_depth=3
)

dt.fit(X_train, y_train)
```

### Explaining the Models
Some models were designed to deal with linear data; others are capable of modeling more complex input data. Some models can be easily interpreted, while others are like black boxes and do not offer much insight into how the prediction is made.

### Regression coefficients
The intercepts and regression coefficients explain the expected value and how the attributes impact the prediction. A positive coefficient indicates that, as the value of an attribute increases, so will the prediction.

### Importance of attributes
Tree-based models from the sciki-learn lib include a *.feature_importances_* attribute to inspect how attributes of a dataset affect the model. We can inspect them numerically, or visualize them on a graph.

### LIME
**LIME** helps explain black box models. It makes a *local* interpretation, rather than a general interpretation, and helps explain a single sample.<br>
For a given point or sample, **LIME** shows which attributes were important in determining the result. It does this by causing a disturbance in the sample in question and adapting a linear model to it.

Example explaining the last sample (which our decision tree predicted would survive) of the training data.
<div class="alert alert-block alert-info">
<b>Tip:</b> LIME doesn't like using DataFrames as input. That's why we convert the data into arrays using the *.values* method.
</div>

```python
explainer = lime_tabular.LimeTabularExplainer(
    X_train.values,
    feature_names=X.columns,
    class_names=["died", "survived"]
)

exp = explainer.explain_instance(
    X_train.iloc[-1].values, dt.predict_proba
)
```
