__pyjanitor__ lib is convenient, but it doesn't allow us to remove whitespace. <br>To do this, we can create a function that transforms everything to lowercase, snake_case, and removes whitespace. After that, we can call the python function passing the dataframe name.

```python
def snake_case_clean(dataframe):
    dataframe.columns = (dataframe.columns
                .str.replace('(?<=[a-z])(?=[A-Z])', '_', regex=True)
                .str.lower()
             )
```

```python
snake_case_clean(df)
df
```
