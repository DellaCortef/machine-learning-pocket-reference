### PCA
Principal Component Analysis uses a matrix (X) of rows (samples) and columns (attributes). It returns a new matrix whose columns are linear combinations of the original columns. These linear combinations maximize variance.<br><br>
Each column is orthogonal to the others (has a right angle), and they are ordered according to decreasing variance.<br><br>
**scikit-learn** has an implementation for this model. It's better to standardize the data before running it. After calling the *.fit* method, we will have access to an attribute *.explained_variance_ratio* that lists the percentage of variance in each column.<br><br>
**PCA** is convenient for visualizing data in two or three dimensions. It is also used as a data pre-processing step to filter random noise. It is suitable for identifying global but not local structures, and works well with linear data.

```python
pca = PCA(random_state=42)
X_pca = pca.fit_transform(
    StandardScaler().fit_transform(X)
)

pca.explained_variance_ratio_
```

```python
pca.components_[0]
```

**Instance parameters**<br>
*n_components=None*
- number of components to be generated. If equal to *None*, returns the same number corresponding to the number of columns<br>

*copy=True*
- will modify data in .fit if True

*whiten=False*
- will clean data after transformation to ensure uncorrelated components

*svd_solver='auto'*
- 'auto' runs SVD 'randomized' if *n_components* is less than 80% of the smallest dimension (faster, but an approximation)

*tol=0.0*
- tolerance for unique values

*iterated_power='auto'*
- number of iterations for svd_solver 'randomized'
<br><br>

**Attributes**

*components_*
- principal components (weight columns of linear combinations for original attributes)

*explained_variance_*
- variance value for each component

*explained_variance_ratio_*
- variance value for each normalized component (sum is 1)

*singular_values_*
- unique values ​​for each component

*mean*
- average of each attribute

*n_components_*
- if *n_components* is a floating point number, this will be the number of components

*noise_variance_*
- estimated noise covariance

The graph of the cumulative sum of the explained variance ratio is called *scree plot*.<br>
This graph shows how much information is stored in the components. The *elbow method* can be used to see if it bends in order to determine how many components to use

```python
fig, ax = plt.subplots(figsize=(6, 4))

ax.plot(pca.explained_variance_ratio_)
ax.set(
    xlabel="Component",
    ylabel="Percent of Explained Variance",
    title="Scree Plot",
    ylim=(0, 1)
)

fig.savefig("images/dimensionality_reduction_pca_scree_plot")
```

Another way to visualize this data is to use a *cumulative chart*. In this example, our original data had 8 columns, but looking at the graph, it appears that we will maintain approximately 90% of the variance if we only use 4 of the **PCA** components

```python
fig, ax = plt.subplots(figsize=(6, 4))

ax.plot(
    np.cumsum(pca.explained_variance_ratio_)
)

ax.set(
    xlabel="Component",
    ylabel="Percent of Explained Variance",
    title="Cumulative Variance",
    ylim=(0, 1)
)

fig.savefig("images/dimensionality_reduction_pca_cumulative_chart")
```

We can use matplotlib's **imshow** function to analyze the impact that attributes have on components. <br>
In this example, the darker the color, the greater the original column's contribution to the component.

```python
fig, ax = plt.subplots(figsize=(6, 4))

plt.imshow(
    pca.components_.T,
    cmap="Spectral",
    vmin=-1,
    vmax=1
)

plt.yticks(range(len(X.columns)), X.columns)
plt.xticks(range(8), range(1, 9))
plt.xlabel("Principal Component")
plt.ylabel("Contribution")
plt.title(
    "Contribution of Features to Components"
)

plt.colorbar()

fig.savefig("images/dimensionality_reduction_pca_contribution_to_components_matplotlib")
```

If we have multiple attributes, we can limit the previous charts by only showing attributes that meet a minimum weight.

```python
comps = pd.DataFrame(
    pca.components_, columns=X.columns
)

min_val = 0.5
num_components = 2
pca_cols = set()

for i in range (num_components):
    parts = comps.iloc[i][
        comps.iloc[i].abs() > min_val
    ]
    pca_cols.update(set(parts.index))

pca_cols
```

We can expand the scatterplot by showing a *loading plot* over it. This graph is called a **biplot** because it contains the scatter plot and the load plot.<br>
Loadings show the strength of attributes and how they are correlated. If they form sharp angles, it is a sign that they are probably correlated. If the angles are 90 degrees, they are likely uncorrelated. Finally, if the angle between them is close to 180 degrees, there will be a negative correlation.

```python
fig, ax = plt.subplots(figsize=(14, 6))

pca_df = pd.DataFrame(
    X_pca,
    columns=[
        f"PC{i+1}"
        for i in range(X_pca.shape[1])
    ],
)

pca_df["status"] = [
    ("deceased", "survived")[i] for i in y
]

evr = pca.explained_variance_ratio_

x_idx = 0 # x_pc
y_idx = 1 # y_pc

ax.set_aspect(evr[y_idx] / evr[x_idx])
x_col = pca_df.columns[x_idx]
y_col = pca_df.columns[y_idx]

sns.scatterplot(
    x=x_col, 
    y=y_col,
    hue="status",
    data=pca_df,
    alpha=0.5,
    ax=ax
)

scale = 8

comps = pd.DataFrame(
    pca.components_, columns=X.columns
)

for idx, s in comps.T.iterrows():
    plt.arrow(
        0,
        0,
        s[x_idx] * scale,
        s[y_idx] * scale,
        color="k"
    )
    plt.text(
        s[x_idx] * scale,
        s[y_idx] * scale,
        idx,
        weight="bold"
    )

fig.savefig("images/dimensionality_reduction_pca_clustering_biplot_seaborn")
```

### UMAP
Uniform Manifold Approximation and projection is a dimensionality reduction technique that uses *manifold learning*. Thus, it tends to put topologically similar items together and seeks to preserve both global and local structure, as opposed to *t-SNE*, which favors local structure.<br><br>
Normalizing the attributes is a good idea to have the values on the same scale. **UMAP** is very sensitive to hyperparameters (*n_neighbors, min_dist, n_components or metric*)

```python
u = umap.UMAP(random_state=42)

X_umap = u.fit_transform(
    StandardScaler().fit_transform(X)
)

X_umap.shape
```

**Instance Parameters**<br>
*n_neighbors=15*
- size of the local neighborhood. Higher values ​​imply a global view, while lower values ​​imply a more local view

*n_components=2*
- number of dimensions for embedding

*metric='euclidean'*
- metric used for distance. It could be a function that accepts two 1D arrays and returns a floating point number

*n_epochs=None*
- number of training epochs. The default will be 200 or 500, depending on the size of the dataset

*learning_rate=1.0*
- learning rate for embedding optimization

*init='sprectral'*
- startup type. Spectral embedding is the default. Can be 'random' or a numpy array of positions

*min_dist=0.1*
- between 0 and 1. Minimum distance between embedded points. Smaller values ​​mean more clumps, higher values ​​mean more dispersion

*spread=1.0*
- determines the distance between points after embedding

*set_op_min_ratio=1.0*
- between 0 and 1: fuzzy union (1) or fuzzy intersection (0)

*local_connectivity=1.0*
- number of neighbors for local connectivity. As this value increases, more local connections will be created

*repulsion_strength=1.0*
- repulsion force. Larger values ​​give more weight to negative samples

*negative_sample_rate=5*
- negative samples per positive sample. Larger values ​​introduce more repulsion, more optimization costs, and more accuracy

*transform_queue_size=4.0*
- aggressiveness in searching for nearest neighbors. A larger value has worse performance but will have greater accuracy

*a=None*
- parameter for embedding control. If equal to *None*, **UMAP** will determine the value from *min_dist* and *spread*

*b=None*
- parameter for embedding control. If equal to *None*, **UMAP** will determine the value from *min_dist* and *spread*

*metric_kwds=None*
- metric dictionary for additional parameters if a function is used for *metric*. Furthermore, *minkowsi* (and others) can be parameterized with this information

*angular_rp_forest=False*
- uses random angular projection

*target_n_neighbors=-1*
- number of neighbors for simplicial sets

*target_metric_kwds=None*
- dictionary of metrics to be used if a function is used in *tager_metric*

*target_metric='categorical'*
- to use supervised reduction. Can also be **L1** or **L2**

*target_weight=0.5*
- weighting factor. Between 0.0 and 1.0, where - means based on data only and 1 means based on target only<br><br>

**Attributes**<br>

*embedding_*
- embedding results

```python
fig, ax = plt.subplots(figsize=(14,6))

pd.DataFrame(X_umap).plot(
    kind="scatter",
    x=0,
    y=1,
    ax=ax,
    c=y,
    alpha=0.2,
    cmap="Spectral"
)

fig.savefig("images/dimensionality_reduction_umap_matplotlib")
```

### t_SNE
The t-SNE (t-Distributed Stochastic Neighboring Embedding) technique is a visualization and dimensionality reduction technique. It uses distributions of input data and embedding to have fewer dimensions, and minimizes the probability of joins between them. As it is very demanding from a processing point of view, it may not be possible to use this technique with a large data set.<br><br>
A characteristic of **t-SNE** is that it is very sensitive to hyperparameters. Furthermore, although it preserves local clusters very well, global information is not preserved. Therefore, the distance between the clusters is not significant. Finally, this is not a deterministic algorithm, and there may be no convergence.<br><br>
Standardizing the data before using the technique is advisable.

```python
X_std = StandardScaler().fit_transform(X)

ts = TSNE()

X_tsne = ts.fit_transform(X_std)
```

**Instance parameters**<br>
*n_components=2*
- number of dimensions for embedding

*perplexity=30.0*
- suggested values ​​are between 5 and 50. Smaller numbers tend to create more rigid agglomerations

*early_exaggeration=12.0*
- controls the rigidity of the clusters and the spacing between them. Larger values ​​mean larger spacing

*learning_rate=200.0*
- in general, between 10 and 1000. If the data resembles a ball, reduce this value. If it looks compressed, elevate it

*n_iter=1000*
- number of iterations

*n_iter_without_progress=300*
- abort if there is no progress after this number of iterations

*min_grad_norm=1e-07*
- optimization will stop if the gradient norm is below this value

*metric='euclidean'*
- distance metric from *scipy.spatial.distance.pdist, pairwise.PAIRWISE_DISTANCE_METRIC* or a function

*init='random'*
- embedding initialization

*method='barnes_hut'*
- gradient calculation algorithm

*angle=0.5*
- for gradient calculation. A value less than 0.2 increases execution time. If it is greater than 0.8, the error will increase<br><br>

**Attributes**<br>

*embedding_*
- embedding vectors

*kl_divergence_*
- Kullback-Leibler divergence

*n_iter_*
- number of iterations

### PHATE
**PHATE** (Potential of Heat-diffusion for Affinity-based Trajectory Embedding) is a tool for visualizing data with many dimensions. It tends to maintain both global structure (like **PCA**) and local structure (like **t-SNE**).<br>

**PHATE** initially encodes local information (points close to each other must remain close). It uses "diffusion" to discover global data and then reduces dimensionality.

```python
p = phate.PHATE(random_state=42)

X_phate = p.fit_transform(X)

X_phate.shape
```

**Instance parameters**<br>

*n_components=2*
- number of dimensions

*knn=5*
- number of neighbors for the kernel. Increase if embedding is disconnected or dataset has more than 100k samples

*decay=40*
- kernel drop rate. Reducing this value increases graph connectivity

*n_landmark=2000*
- landmarks to be used

*t='auto'*
- diffusion power. A smoothing is done on the data. Increase this value if the embedding lacks structure. Decrease if the structure is rigid and compact

*gamma=1*
- logarithmic potential (between -1 and 1). If the embeddings are concentrated around a single point, try setting this value to 0

*n_pca=100*
- number of main components for neighborhood calculation

*knn_dist='euclidian'*
- KNN metric

*mds_dist='euclidian'*
- MDS (multidimensional scaling) metric

*mds='metric'*
- MDS algorithm for dimension reduction

*n_jobs=1*
- number of CPUs to be used<br><br>


**Attributes**<br>

*X*
- input data

*embedding*
- embedding space

*diff_op*
- broadcast operator

*graph*
- KNN graph built from input data

```python
fig, ax = plt.subplots(figsize=(6, 4))

phate.plot.scatter2d(p, c=y, ax=ax, alpha=0.5)

fig.savefig("images/dimensionality_reduction_phate_matplotlib")
```
