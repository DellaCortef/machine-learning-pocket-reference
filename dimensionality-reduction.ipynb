```python
def tweak_titanic(df):
    df = df.drop(columns = ['name',
                        'name_wiki',
                        'wiki_id',
                        'hometown',
                        'destination',
                        'ticket',
                        'lifeboat',
                        'body',
                        'cabin',
                        'age'
        ]
    ).pipe(pd.get_dummies, drop_first=True)
    return df

def get_train_test_X_y(
    df, y_col, size=0.3, std_cols=None
):
    y = df[y_col]
    X = df.drop(columns=y_col)
    X_train, X_test, y_train, y_test = model_selection.train_test_split(
        X, y, test_size=size, random_state=42
    )
    cols = X.columns
    num_cols = [
        "pclass",
        "age_wiki",
        "sib_sp",
        "parch",
        "fare",
    ]
    fi = impute.IterativeImputer()

    fitted = fi.fit_transform(X_train[num_cols])
    X_train = X_train.assign(**{c:fitted[:,i] for i, c in enumerate(num_cols)})
    test_fit = fi.transform(X_test[num_cols])
    X_test = X_test.assign(**{c:test_fit[:,i] for i, c in enumerate(num_cols)})
    if std_cols:
        std = preprocessing.StandardScaler()
        fitted = std.fit_transform(X_train[std_cols])
        X_train = X_train.assign(**{c:fitted[:,i] for i, c in enumerate(std_cols)})
        test_fit = std.transform(X_test[std_cols])
        X_test = X_test.assign(**{c:test_fit[:,i] for i, c in enumerate(std_cols)})

    return X_train, X_test, y_train, y_test

ti_df = tweak_titanic(df)
std_cols = "pclass,age_wiki,sib_sp,fare".split(",")
X_train, X_test, y_train, y_test = get_train_test_X_y(
    ti_df, "survived", std_cols=std_cols
)


X = pd.concat([X_train, X_test])
y = pd.concat([y_train, y_test])
```

### PCA
Principal Component Analysis uses a matrix (X) of rows (samples) and columns (attributes). It returns a new matrix whose columns are linear combinations of the original columns. These linear combinations maximize variance.<br><br>
Each column is orthogonal to the others (has a right angle), and they are ordered according to decreasing variance.<br><br>
**scikit-learn** has an implementation for this model. It's better to standardize the data before running it. After calling the *.fit* method, we will have access to an attribute *.explained_variance_ratio* that lists the percentage of variance in each column.<br><br>
**PCA** is convenient for visualizing data in two or three dimensions. It is also used as a data pre-processing step to filter random noise. It is suitable for identifying global but not local structures, and works well with linear data.

```python
pca = PCA(random_state=42)
X_pca = pca.fit_transform(
    StandardScaler().fit_transform(X)
)

pca.explained_variance_ratio_
```

```python
pca.components_[0]
```

**Instance parameters**<br>
*n_components=None*
- number of components to be generated. If equal to *None*, returns the same number corresponding to the number of columns<br>

*copy=True*
- will modify data in .fit if True

*whiten=False*
- will clean data after transformation to ensure uncorrelated components

*svd_solver='auto'*
- 'auto' runs SVD 'randomized' if *n_components* is less than 80% of the smallest dimension (faster, but an approximation)

*tol=0.0*
- tolerance for unique values

*iterated_power='auto'*
- number of iterations for svd_solver 'randomized'
<br><br>

**Attributes**

*components_*
- principal components (weight columns of linear combinations for original attributes)

*explained_variance_*
- variance value for each component

*explained_variance_ratio_*
- variance value for each normalized component (sum is 1)

*singular_values_*
- unique values ​​for each component

*mean*
- average of each attribute

*n_components_*
- if *n_components* is a floating point number, this will be the number of components

*noise_variance_*
- estimated noise covariance

The graph of the cumulative sum of the explained variance ratio is called *scree plot*.<br>
This graph shows how much information is stored in the components. The *elbow method* can be used to see if it bends in order to determine how many components to use

```python
fig, ax = plt.subplots(figsize=(6, 4))

ax.plot(pca.explained_variance_ratio_)
ax.set(
    xlabel="Component",
    ylabel="Percent of Explained Variance",
    title="Scree Plot",
    ylim=(0, 1)
)

fig.savefig("images/dimensionality_reduction_pca_scree_plot")
```

Another way to visualize this data is to use a *cumulative chart*. In this example, our original data had 8 columns, but looking at the graph, it appears that we will maintain approximately 90% of the variance if we only use 4 of the **PCA** components

```python
fig, ax = plt.subplots(figsize=(6, 4))

ax.plot(
    np.cumsum(pca.explained_variance_ratio_)
)

ax.set(
    xlabel="Component",
    ylabel="Percent of Explained Variance",
    title="Cumulative Variance",
    ylim=(0, 1)
)

fig.savefig("images/dimensionality_reduction_pca_cumulative_chart")
```

We can use matplotlib's **imshow** function to analyze the impact that attributes have on components. <br>
In this example, the darker the color, the greater the original column's contribution to the component.

```python
fig, ax = plt.subplots(figsize=(6, 4))

plt.imshow(
    pca.components_.T,
    cmap="Spectral",
    vmin=-1,
    vmax=1
)

plt.yticks(range(len(X.columns)), X.columns)
plt.xticks(range(8), range(1, 9))
plt.xlabel("Principal Component")
plt.ylabel("Contribution")
plt.title(
    "Contribution of Features to Components"
)

plt.colorbar()

fig.savefig("images/dimensionality_reduction_pca_contribution_to_components_matplotlib")
```

If we have multiple attributes, we can limit the previous charts by only showing attributes that meet a minimum weight.

```python
comps = pd.DataFrame(
    pca.components_, columns=X.columns
)

min_val = 0.5
num_components = 2
pca_cols = set()

for i in range (num_components):
    parts = comps.iloc[i][
        comps.iloc[i].abs() > min_val
    ]
    pca_cols.update(set(parts.index))

pca_cols
```
