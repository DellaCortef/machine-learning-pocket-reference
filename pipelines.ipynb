**scikit-learn** uses the notion of pipeline. By using the *Pipeline* class, we can not only chain transformers and models, but also treat the entire process as a single **scikit-learn** model. It is even possible to use custom logic.

### Classification Pipeline

```python
class TitanicTransformer(
    BaseEstimator, TransformerMixin
):
    def transform(self, X):
        # assume that X is the result of reading an excel file
        X = X.drop(columns="survived")
        return X
    
    def fit(self, X, y):
        return self
    
pipe = Pipeline(
    [
        ("titan", TitanicTransformer()),
        ("impute", impute.IterativeImputer()),
        (
            "std",
            preprocessing.StandardScaler(),
        ),
        ("rf", RandomForestClassifier()),
    ]
)
```

```python
X_train2, X_test2, y_train2, y_test2 = train_test_split(
    df,
    df.survived,
    test_size=0.3,
    random_state=42
)
```

```python
pipe.fit(X_train2, y_train2)
```

```python
pipe.score(X_test2, y_test2)
```

Pipelines can be used in grid search. Our *param_grid* must have the parameters prefixed by the name of the pipeline step, followed by two underscores.

```python
params = {
    "rf__max_features": [0.4, "auto"],
    "rf__n_estimators": [15, 200]
}
```

```python
grid = model_selection.GridSearchCV(
    pipe, cv=3, param_grid=params
)

grid.fit(df, df.survived)
```

With this, we can extract the best parameters and train the final model.

```python
grid.best_params_
```

```python
pipe.set_params(**grid.best_params_)
```

```python
pipe.fit(X_train2, y_train2)
```

```python
pipe.score(X_test2, y_test2)
```
